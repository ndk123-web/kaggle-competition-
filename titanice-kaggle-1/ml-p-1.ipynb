**1 Import Libraries**

```python
import numpy as np 
import pandas as pd
print("IMPORTED LIBRARIES")
```

**2 Load Data**

```python
train = pd.read_csv("/kaggle/input/titanic/train.csv")
test = pd.read_csv("/kaggle/input/titanic/test.csv")

print("Loaded Train And Test Model")
```

**3 Look At The Data**

```python
train.head()
```

```python
train["Survived"].value_counts()
```

**4 Keep Important Features**

```python
features = ["Pclass", "Sex", "Age"]
test.head()
```

**5 Convert Male And Female To Numeric Value** *Encoding***

```python
train["Sex"] = train["Sex"].map({"male": 0, "female": 1})
test["Sex"] = test["Sex"].map({"male": 0, "female": 1})

test.head()
```

**6 Fill NA Values**


```python
train["Age"] = train["Age"].fillna(train["Age"].median())
test["Age"] = test["Age"].fillna(test["Age"].median())

train["Pclass"] = train["Pclass"].fillna(train["Pclass"].median())
test["Pclass"] = test["Pclass"].fillna(test["Pclass"].median())

train["Fare"] = train["Fare"].fillna(train["Fare"].median())
test["Fare"] = test["Fare"].fillna(test["Fare"].median())

print("HANDLED AGE NULL VALUES")
```

**7 Load X AND Y which is Input And Output**


```python
X = train[features]
Y = train["Survived"]

print("LOAD X AND Y")
```

**8 Create A Logistic Regression Model**

```python
from sklearn.linear_model import LogisticRegression , BayesianRidge
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

# model = LogisticRegression()
# model = BayesianRidge()
# model = GaussianNB()
model = DecisionTreeClassifier()
model.fit(X,Y)

print("MODEL TRAINEDD")
```

**9 Make Predictions**

```python
X_test = test[features]
predictions = model.predict(X_test)
```

**10 Create CSV For Predictions**

```python
submission = pd.DataFrame(
    {
        "PassengerId":test["PassengerId"],
        # "Sex":train["Sex"],
        "Survived": predictions
    }
)

submission.to_csv("Bayesian-Ridge-submission-test-1.csv",index = False)
```
